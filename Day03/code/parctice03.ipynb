### https://maoyan.com/board/4

```python
import requests
import re
```

```python
url = 'https://maoyan.com/board/4'
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36'}
```

```python
response = requests.get(url, headers = header)
print(response.status_code)
```

```python
print(response.text)
```

```python
title = re.findall(r'data-val="{movieId:.*?}">(.*?)</a>', response.text)
print(title)
```

```python
star = re.findall(r'<p class="star">(.*?)</p>', response.text, re.S)
print(star)
```

```python
time = re.findall(r'<p class="releasetime">(.*?)</p>', response.text)
print(time)
```

```python
for i in range(len(title)):
    print(i+1, '\t' + title[i], '\t' + star[i].strip(), '\t' + time[i])
```

```python
image_url = re.findall(r'https://p0.meituan.net/movie/(.*?).jpg', response.text)
print(image_url)
```

```python
for i in image_url:
    img = requests.get('https://p0.meituan.net/movie/' + i + '.jpg', headers = header)
    # print(img.status_code)
    with open('image/%s.jpg'%(i), 'wb') as file:
        file.write(img.content)
        print('%s is downloaded!'%(i))
```
